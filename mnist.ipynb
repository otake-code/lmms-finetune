```python
import os
import json

def create_jsonl(root_dir, output_dir, output_filename):
    """
    root_dir: '/dataset/mnist/train' または '/dataset/mnist/test'
    output_dir: 'jsons/mnist'
    output_filename: 'mnist_train.jsonl' or 'mnist_test.jsonl'
    """
    os.makedirs(output_dir, exist_ok=True)
    records = []
    for label_str in ['0', '1']:
        dir_path = os.path.join(root_dir, label_str)
        if not os.path.isdir(dir_path):
            continue
        for fname in sorted(os.listdir(dir_path)):
            if not (fname.endswith('.png') or fname.endswith('.jpg') or fname.endswith('.jpeg')):
                continue
            img_path = os.path.join(dir_path, fname)
            file_num, _ = os.path.splitext(fname)
            prompt = (
                "<image>\n"
                "Which digit is shown in this image? Respond with the digit."
            )
            answer = label_str
            record = {
                "source": file_num,
                "images": [img_path],
                "videos": [],
                "system_prompt": None,
                "conversations": [prompt, answer]
            }
            records.append(record)
    output_path = os.path.join(output_dir, output_filename)
    with open(output_path, 'w', encoding='utf-8') as f:
        for rec in records:
            f.write(json.dumps(rec, ensure_ascii=False) + '\n')
    return output_path, len(records)

if __name__ == '__main__':
    train_path, train_count = create_jsonl(
        '/dataset/mnist/train',
        'jsons/mnist',
        'mnist_train.jsonl'
    )
    test_path, test_count = create_jsonl(
        '/dataset/mnist/test',
        'jsons/mnist',
        'mnist_test.jsonl'
    )
    print(f"{train_path}: {train_count} 件")
    print(f"{test_path}:  {test_count} 件")
```
